```python
score = np.loadtxt("/content/survey.txt", dtype=int)
score
```

```python
num_feedback_recorded = score.shape[0]
num_feedback_recorded
```

```python
num_feedback_recorded = len(score)
num_feedback_recorded
```

```python
num_feedback_recorded = score.size # Number of elements in the array.
num_feedback_recorded
```

```python
promoters_mask = score>=9
promoters = score[promoters_mask]
num_promoters = promoters.shape[0]

per_promoters = num_promoters*100/num_feedback_recorded
per_promoters
```

```python
detractor_mask = score<=6
detractors = score[detractor_mask]
num_detractors = detractors.shape[0]

per_detractors = num_detractors*100/num_feedback_recorded
per_detractors
```

```python
NPS = per_promoters-per_detractors
print(NPS)
```

```python
print(np.round(NPS))
```

```python
print(np.round(NPS,2))
```
